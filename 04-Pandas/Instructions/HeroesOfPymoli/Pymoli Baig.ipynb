```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file)
df.head()
```

# Player Count

```python
#number of players
players = df.SN.nunique()

playersTable = pd.DataFrame()

playersTable["Total Players"] = [players]

playersTable
```

# Total Purchasing Analysis

```python
#info in table
diffItems = df["Item ID"].nunique()
avgPrice = df.Price.mean()
purchases = len(df)
revenue = df.Price.sum()

#create table
purchasingAnalysis = pd.DataFrame()
purchasingAnalysis["# of Unique Items"] = [diffItems]
purchasingAnalysis["Average Price"] = [avgPrice]
purchasingAnalysis["# of Purchases"] = [purchases]
purchasingAnalysis["Total Revenue"] = [revenue]

#format
purchasingAnalysis["Average Price"] = purchasingAnalysis["Average Price"].map("${:,.2f}".format)
purchasingAnalysis["Total Revenue"] = purchasingAnalysis["Total Revenue"].map("${:,.2f}".format)

purchasingAnalysis 
```

# Demographics - Gender

```python
#stackoverflow
diffPeople_df = df.groupby(['SN','Gender']).size().reset_index().rename(columns={0:'count'})
diffPeople_df
```

```python
#columns
countGender = diffPeople_df.Gender.value_counts()
percentGender = countGender / len(diffPeople_df)

#table
genderTable = pd.DataFrame()
genderTable["Total Count"] = countGender
genderTable["Percentage of Players"] = percentGender * 100

genderTable["Percentage of Players"] = genderTable["Percentage of Players"].map("{:.2f}%".format)

genderTable
```


# Gender Purchasing Analysis


```python
 #columns
genSize = df.groupby("Gender").size()
genAvgPrice = df.groupby("Gender").Price.mean()
genTotPrice = df.groupby("Gender").Price.sum()
genAvgPerPerson = df.groupby("Gender").Price.sum() / genderTable["Total Count"]

#table
genderPurchasingTable = pd.DataFrame()
genderPurchasingTable["Purchase Count"] = genSize
genderPurchasingTable["Avg Purchase Price"] = genAvgPrice
genderPurchasingTable["Total Purchase Value"] = genTotPrice
genderPurchasingTable["Avg Total Purchase per Person"] = genAvgPerPerson


genderPurchasingTable["Avg Purchase Price"] = genderPurchasingTable["Avg Purchase Price"].map("${:,.2f}".format)
genderPurchasingTable["Total Purchase Value"] = genderPurchasingTable["Total Purchase Value"].map("${:,.2f}".format)
genderPurchasingTable["Avg Total Purchase per Person"] = genderPurchasingTable["Avg Total Purchase per Person"].map("${:,.2f}".format)


genderPurchasingTable
```

# Demographics - Age

```python
# Create bins for values based on player ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

# Create bin labels
group_labels = [" <10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice data into bins
df["Age Group"] = pd.cut(df["Age"], bins, labels=group_labels).astype(str)
df.head()
```

```python
people2_df = df.groupby(['SN','Age Group']).size().reset_index().rename(columns={0:'count'})
people2_df
```

```python
ageTable = people2_df.groupby("Age Group").size().reset_index().rename(columns={0:'Total Count'})
ageTable["% of Players"] = ageTable["Total Count"] / len(people2_df) * 100

ageTable["% of Players"] = ageTable["% of Players"].map("{:.2f}%".format)

ageTable = ageTable.set_index("Age Group")
ageTable
```

# Age Purchasing Analysis

```python
#columns
ageSize = df.groupby("Age Group").size()
ageAvgPrice = df.groupby("Age Group").Price.mean()
ageTotPrice = df.groupby("Age Group").Price.sum()
ageAvgPerPerson = df.groupby("Age Group").Price.sum() / ageTable["Total Count"] 

#table
agePurchasingTable = pd.DataFrame()
agePurchasingTable["Purchase Count"] = ageSize
agePurchasingTable["Avg Purchase Price"] = ageAvgPrice
agePurchasingTable["Total Purchase Value"] = ageTotPrice
agePurchasingTable["Avg Total Purchase per Person"] = ageAvgPerPerson

agePurchasingTable["Avg Purchase Price"] = agePurchasingTable["Avg Purchase Price"].map("${:,.2f}".format)
agePurchasingTable["Total Purchase Value"] = agePurchasingTable["Total Purchase Value"].map("${:,.2f}".format)
agePurchasingTable["Avg Total Purchase per Person"] = agePurchasingTable["Avg Total Purchase per Person"].map("${:,.2f}".format)

agePurchasingTable
```


# Top Spenders

```python
#columns
snSize = df.groupby("SN").size()
snAvgPrice = df.groupby("SN").Price.mean()
snTotPrice = df.groupby("SN").Price.sum()

#table
snPurchasingTable = pd.DataFrame()
snPurchasingTable["Purchase Count"] = snSize
snPurchasingTable["Avg Purchase Price"] = snAvgPrice
snPurchasingTable["Total Purchase Value"] = snTotPrice

snPurchasingTable["Avg Purchase Price"] = snPurchasingTable["Avg Purchase Price"].map("${:,.2f}".format)
snPurchasingTable["Total Purchase Value"] = snPurchasingTable["Total Purchase Value"].map("${:,.2f}".format)

snPurchasingTable = snPurchasingTable.sort_values(by="Total Purchase Value", ascending=False)
snPurchasingTable.head()
```

# Most Popular Items

```python
#columns
itemSize = df.groupby(["Item ID", "Item Name"]).size()
itemAvgPrice = df.groupby(["Item ID", "Item Name"]).Price.mean()
itemTotPrice = df.groupby(["Item ID", "Item Name"]).Price.sum()

#table
itemPurchasingTable = pd.DataFrame()
itemPurchasingTable["Purchase Count"] = itemSize
itemPurchasingTable["Item Price"] = itemAvgPrice
itemPurchasingTable["Total Purchase Value"] = itemTotPrice

itemPurchasingTable = itemPurchasingTable.sort_values(by="Purchase Count", ascending=False)

itemPurchasingTable["Item Price"] = itemPurchasingTable["Item Price"].map("${:,.2f}".format)
itemPurchasingTable["Total Purchase Value"] = itemPurchasingTable["Total Purchase Value"].map("${:,.2f}".format)

itemPurchasingTable.head()
```
